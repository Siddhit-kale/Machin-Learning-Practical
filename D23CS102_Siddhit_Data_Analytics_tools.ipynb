5) Adding annotations

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,15)
y = x ^ 3
z = x ^ 3
t = x ^ 4
# Labling the Axes and title
plt.title("Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x , y)
plt.annotate(xy = [2,1], text="Second entry")
plt.annotate(xy = [4,6], text="Third entry")
```

6) Adding Legend

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10)
y = x ^ 2
z = x ^ 3
t = x ^ 4
# Labling the Axes and title
plt.title("Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x , y)
plt.annotate(xy = [2,1], text="Second entry")
plt.annotate(xy = [4,6], text="Third entry")

# Adding Legends
plt.plot(x, z)
plt.plot(x, t)
plt.legend(['Race1', 'Race2', 'Race3'], loc=4)
```

7) SUperimposing 2 Graph

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.plot(t1, f(t1), '+', t2, f(t2), "gold")
```

8) Plotting Pie Graph

```python
import matplotlib.pyplot as plt
labels = 'Forgs', 'Hogs', 'Dogs', 'Logs'
sizes = [25, 25, 25, 25]
explode = ( 0, 0, 0.2, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle = 90)
ax1.axis('equal')
plt.legend(['Forgs', 'Hogs', 'Dogs', 'Logs'], loc=0)
plt.show()
```

10) Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(0, 5, 11)
y = x ** 2
plt.subplot(1, 2, 1)
plt.plot(x ,x ,'r--')
plt.subplot(1, 2 ,2)
plt.plot(y, x, 'g*-')
plt.xlabel("X Axis")
plt.ylabel("Y Axis Title here")
plt.title("String title here")
```

11) Another way of using Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots()

#genrate 2 objects, fig and ax
ax.plot(x, x*2, "b.-")
ax.plot(x, x**3, "r<")
ax.plot(x, x**4, "y>")
ax.plot(x, x**5, "g<")
ax.plot(x, x**6, "c*-")

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("String title here")
```

12) Histogram With text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins,patches = plt.hist(x, 50, density=1, facecolor='gold', alpha=0.7)
plt.xlabel('smarts')
plt.ylabel('probability')
plt.title('Histogram Of IQ')
plt.text(60, .035, r'$\mu=100, \ \ sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

13) plot different varity of graph as subplots

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

n = np.array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)
fig, axes = plt.subplots(1, 4, figsize=(12,3))
axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(xx, xx**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(xx, xx**2, xx**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

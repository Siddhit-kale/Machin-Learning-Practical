```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load dataset 
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)
print(california_housing.DESCR)
```

```python
housing = fetch_california_housing()

data = pd.DataFrame(housing.data,columns=housing.feature_names)
data.head()
data['PRICE'] = housing.target
data.head(5)
```

```python
print(housing.feature_names)
```

```python
# features (input variables)
x = data.drop('PRICE', axis=1)
# axis = 0 for index and axis = 1 for column

# Target variables
y = data['PRICE']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
x_train.head(10)
```

```python
model = LinearRegression()

# train the model
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# calculate mean sequared error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean Sequared Error:", mse)

R2 = r2_score(y_test, y_pred)
print("R-squared:", R2)
```

```python
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted price')
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
plt.plot(y_test, x_test,c="red")
plt.plot(y_pred, x_test, c="blue")
plt.xlabel("Features", fontsize=10)
plt.ylabel("Predicted price and Actual price", fontsize=10)
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/sample_submission.csv') # Give path of the Dataset
df
```

```python
plt.xlabel('Id')
plt.ylabel('SalePrice')
plt.scatter(df.Id, df.SalePrice, color='gold', marker='+')

new_df  = df.drop('SalePrice', axis='columns')
SalePrice = df.SalePrice

# x = df.drop('unname:0', axis=1)
# x

# Create Linear regression object
req = linear_model.LinearRegression()
req.fit(new_df, SalePrice)
req.predict([[5]])
```

```python
plt.xlabel('Id')
plt.ylabel('SalePrice')
plt.scatter(df.Id, df.SalePrice, color='red', marker='+')
plt.plot(df.Id, req.predict(new_df), color='blue')
```
